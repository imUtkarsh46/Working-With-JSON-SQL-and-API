# Working with API

```python
import pandas as pd
import requests
```

```python
res = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=d9e992aef7a2a379ac9756124788aa0b&language=en-US&page=1')
```

```python
res
```

```python
pd.DataFrame(res.json()['results'])[['id', 'original_title','popularity', 'release_date','vote_average','vote_count' ]]
```

```python
data = pd.DataFrame()
for i in range(1, 507):
    res = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=d9e992aef7a2a379ac9756124788aa0b&language=en-US&page={i}')
    temp_df = pd.DataFrame(res.json()['results'])[['id', 'original_title','popularity', 'release_date','vote_average','vote_count' ]]
    data = data.append(temp_df)
```

```python
data
```

```python
data.to_csv('Movies 2022.csv')
```
